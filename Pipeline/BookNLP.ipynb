```python
model_params = {
		"pipeline":"entity,quote,supersense,event,coref",
		"model":"big",
	}

booknlp= BookNLP("en", model_params)
```

```python
INPUT_FOLDER = "/content/drive/MyDrive/booknlp_pipeline/input"
OUTPUT_FOLDER = "/content/drive/MyDrive/booknlp_pipeline/output"

while True:
    files = [f for f in os.listdir(INPUT_FOLDER) if f.endswith(".txt")]
    if files:
        for file in files:
            input_path = os.path.join(INPUT_FOLDER, file)
            output_dir = os.path.join(OUTPUT_FOLDER, file.replace(".txt", ""))
            book_id = os.path.splitext(file)[0]

            # run BookNLP
            booknlp.process(input_path, output_dir, book_id)

            # optionally remove input file
            os.remove(input_path)

        break
    else:
        print("⏳ No input found. Retrying...")
        time.sleep(10)
```
